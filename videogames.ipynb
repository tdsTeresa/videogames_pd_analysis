```python
# Conversión de tipo object a 'strings'
df['Name']= df['Name'].astype('string')
df['Platform']= df['Platform'].astype('string')
df['Genre']= df['Genre'].astype('string')
df['Publisher']= df['Publisher'].astype('string')
```

```python
# Búsqueda de nulos
df.isnull().sum()
```

```python
# Valores nulos en Año, algunos videojuegos tienen el año de lanzamiento en el nombre
# Algunos valores faltantes en Publisher están etiquetados como Unknown
year_null= df[df['Year'].isnull()]
year_null
```

```python
# Expresión regular que busca números de 4 cifras en los nombres del df de años nulos
year_null_df= year_null[(year_null['Name'].str.contains(r'\d{4}'))]
year_null_df
```

```python
# Diccionario de index y años que se usará para reemplazar nulos en el df principal
years= {179: 2004, 377: 2004, 470: 2006, 1649: 2003, 2586: 2009, 3501: 2002, 4797: 2003, 5162: 2003, 5240: 2013, 5669: 2005, 5901: 2003, 8929: 2005, 12922: 2011, 13870: 2006, 15739: 2007, 15865: 2008, 16057: 2008}
```

```python
# Búsqueda de nulos en Publisher
df[df['Publisher'].isnull()]
```

```python
# Matriz de correlación del df principal
df_corr= df.corr(numeric_only=True)
df_corr
```

```python
# Definimos la mitad inferior de la correlación con np.tril y sus dimensiones
df_heatmap= df_corr.where(np.tril(np.ones(df_corr.shape).astype(bool)))
```

```python
# Importamos la librería seaborn para el heatmap y matplotlib para insertar un título
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df_heatmap, annot= True, cmap='Blues')
plt.title('Heatmap de ventas')
plt.show()
```

#### Interpretación:
- Poca correlación entre los años y las ventas.

- Correlación fuerte entre ventas en Norteamérica y globales.   
        <b>Las ventas de Norteamérica influyen fuertemente en las ventas globales.</b><br><br>
- Correlación alta entre Europa y las ventas globales.  
        <b>Europa también contribuye grandemente en las ventas globales.</b><br><br>
- Correlación fuerte de las otras regiones y las ventas globales.

```python
# Géneros con mayor y menor crecimiento en los últimos años:
# Para este análisis se toman datos de 2012 a 2016, ya que desde el año 2017 hay muy pocos juegos en comparación con otros años y un salto hasta 2020
df_pct= df_years.loc[df_years['Year'].between(2012, 2016)].sort_values(by= 'Global_Sales', ascending=False)
df_pct
```

```python
# Pivot table del df filtrado, teniendo la suma de ventas globales, el año y los géneros
df_pct_pivot= pd.pivot_table(df_pct, values= 'Global_Sales', index= ['Year'], columns= ['Genre'], aggfunc= sum).fillna(0).reset_index()
```

```python
df_pct_pivot
```

```python
# Ajuste del df y cálculo del porcentaje de ventas respecto al año anterior en cada género
genre_pct= df_pct_pivot[list(df_pct_pivot.columns)[1:13]].pct_change()
```

```python
# Se define un estilo de fondo
genre_pct.style.background_gradient(cmap= 'Blues')
```

```python
# Media de los porcentajes de ventas de mayor a menor
genre_pct.mean().sort_values(ascending= False)
```

### Crecimiento de géneros de videojuegos en los últimos 4 años

- Géneros con menor crecimiento: <b>Simulación y Puzzle.</b>
